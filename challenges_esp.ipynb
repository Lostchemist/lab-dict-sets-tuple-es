#### En la celda de abajo, crea una lista llamada `sample_list_1` con 80 valores aleatorios.

Requisitos:

* Cada valor es un entero entre 0 y 100.
* Cada valor en la lista es √∫nico.

Imprime `sample_list_1` para revisar sus valores.

*Sugerencia: usa `random.sample` ([referencia](https://docs.python.org/3/library/random.html#random.sample)).*

````python
sample_list_1 = random.sample(range(100), k=80)

#Check if all the numbers in the list are unique
seen_numbers = set()
repetitive = False

for number in sample_list_1:
    if number in seen_numbers:
        repetitive = True
        break
    else:
        seen_numbers.add(number)

if repetitive:
    print("There are repetitive numbers in sample_list_1.")
else:
    print("There are no repetitive numbers in sample_list_1.")
    
print(sample_list_1)
print("\n Sample list has ",len(sample_list_1), "random elements between 0 and 100")
````
